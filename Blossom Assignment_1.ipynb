# Melting the data

```python
bm = pd.melt(bill,id_vars = ['artiste', 'track', 'genre', 'date.entered', 'date.peaked'],var_name='week',value_name='rank')
```

```python
bm
```

# Checking for missing values

```python
bm.isna().sum()
```

```python
bm = bm.dropna(thresh=7)
```

```python
print(bm.shape)
```

```python
bm.dtypes
```

```python
bm['date.entered'] = pd.to_datetime(bm['date.entered'])
bm['date.peaked'] = pd.to_datetime(bm['date.peaked'])
```

```python
bm.dtypes
```

```python
bm['entered_to_peak'] =bm['date.peaked']- bm['date.entered']
```

```python
bm.shape
```

```python
bm.dtypes
```

```python
bm['entered_to_peak']=bm['entered_to_peak'].dt.days
bm['entered_to_peak']
```
